```python
x = xy_data.iloc[:, 0]  # use iloc to get first column
y = xy_data.iloc[:, 1]  # use iloc to get second column

y_model = model.__ * x + model.__  # calculate the predictions for y

residuals = y_model - y  # residuals are just the difference between y_model and y

rmse = np.sqrt(np.mean(__**2))  # simply the square root of the mean squared error

# these are equivalent definitions of R^2:
r2 = 1 - np.sum((y_model-y)**2) / np.sum((np.mean(y) - y)**2)
r2 = np.var(y_model) / np.var(y)  # the "explained variance" definition

print(f'RMSE = {rmse}, R^2 = {r2}')  # print the result
```
